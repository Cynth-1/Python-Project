Size of Objects in Memory

```python
# An integer is 24bytes
sys.getsizeof(1)
```

```python
# Longs are larger
sys.getsizeof(10**100)
```

```python
# Numpy is much smaller
np.dtype(int).itemsize
```

```python
# List are larger
sys.getsizeof([1])
```

```python
# An array of one element in numpy
np.array([1]).nbytes
```
